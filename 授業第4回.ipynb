```python
import pandas as pd
#問15
df_iris = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/【授業資料】フォルダ/授業第4回/iris.csv")
print(df_iris)

print("\n"+"Classのユニーク値とその出現回数"+"\n")
print(df_iris['Class'].value_counts())
```
