### コード4-23 データ準備・前処理からモデルの評価までの全体像

```python
import pandas as pd


#データの読み込み
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/【授業資料】フォルダ/授業第5回/KvsT.csv")


#特徴量と正解データに分割
xcol = ["身長","体重","年代"]
x = df[xcol]
t = df["派閥"]

#モデルの準備と学習
from sklearn import tree
model = tree.DecisionTreeClassifier(random_state=0)
model.fit(x, t)

model.score(x,t)
```
